```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv")
df
```

```python
import altair as alt

brush = alt.selection_interval(encodings=['x', 'y'])

dropdown = alt.binding_select(options=["Unusual", "Storage", "Education"])
usage_status = alt.selection_single(fields=['Usage Description'], bind=dropdown, name="Usage Description")

chart1 = alt.Chart.from_dict({
    "data": {"url": "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"},
    "mark": {"type": "circle", "filled": True},
    "height": 300,
    "width": 500,
    "encoding": {
        "x": {"field": "Year Constructed", "type": "temporal"},
        "y": {"field": "Total Floors", "type": "quantitative"},
        "color": {"field": "Total Floors", "type": "quantitative"},
        "tooltip": [{"field": "Square Footage", "type": "quantitative"}]
    }
}).add_selection(brush, usage_status)

final_bar = chart1.transform_filter(usage_status)

final_bar
```

```python
chart2 = alt.Chart.from_dict({
    "data":{"url":"https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"},
  "mark":"line",
  "encoding":{
    "x":{"field":"Total Floors", "type":"quantitative"},
    "y":{"aggregate":"mean", "field":"Square Footage"}
  }
})
```
